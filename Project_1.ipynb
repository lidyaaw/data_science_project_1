### Group `Churn` and `MonthlyCharges`, and classify the `MonthlyCharges` value into 2 class utilizing average
If the value of `MonthlyCharges` higher than the average then this data is classified as `above`, whereas the `MonthlyCharges` value lower than the average value classified as `below`.

```python
#to sort the data value descendingly and show the first 10 rows
df.sort_values('MonthlyCharges', ascending= False)[:10]
```

```python
#to check the average value in monthlycharges
MC_MEAN = df['MonthlyCharges'].mean()
MC_MEAN
```

```python
#to create a function which will define the value as above or below if it is bigger or smaller than the average value in monthlycharges.
def monthly_charges_status(charge):
  if charge >= MC_MEAN:
    return'above'
  else:
    return 'below'
#to apply the function that we've made to a new column.
df['MonthlyChargesStatus'] = df['MonthlyCharges'].apply(monthly_charges_status)
```

```python
df.head()
```

### A significant difference spotted 
If the monthly charges status is **above** the average.

The <u>higher the transaction **above the average**</u> made by people, the <u>higher the chance of that customer **becoming churn**.</u>
ðŸ‘‡ðŸ‘‡ðŸ‘‡ðŸ‘‡ðŸ‘‡

```python
#to check the percentage of Churn based on the monthly charges status
df.groupby(['MonthlyChargesStatus', 'Churn']).size() / tot_data * 100
```

### Here we are planning to group `Churn` and `duration`, and classify the `duration` into several classes
`duration` is the number of month that user have use our services, i.e. their `TotalCharges` divided by `MonthlyCharges`

Then we classify the `duration` into several classes, by using the quantile: q<sub>1</sub> (25 %), q<sub>2</sub> (50 %) and q<sub>3</sub> (75 %). 

We will classify as 

`very low` for customer that has duration **lower** than q<sub>1</sub>,

`low` for customer that has duration **lower** than q<sub>2</sub>, 

`medium` for customer that has duration **lower** than q<sub>3</sub>, 

`high` for customer that has duration **higher** than q<sub>3</sub>.

#### First we need to get the duration of each customer

```python
#to find the duration of months, the user have been doing transaction
df['duration'] = df['TotalCharges'] / df['MonthlyCharges']
df.head()
```

```python
df.describe()
```

#### Then we classify them using a python function

```python
#25% user is below 8.7 (9) months
#50% user is below 28.6 (29) months
#75% user is below 55 months
def duration_status(duration):
  if duration <= 9: return "very low"
  elif duration <= 29: return 'low'
  elif duration <= 55: return 'medium'
  else: return "high"

df['DurationStatus'] = df['duration'].apply(duration_status)
```

```python
df.head()
```

### A significant difference spotted 
If the duration status is **very low**.

The <u>**lower** the duration</u> people charges/use our services, the <u>higher the chance of that customer **becoming churn**.</u>
ðŸ‘‡ðŸ‘‡ðŸ‘‡ðŸ‘‡ðŸ‘‡

```python
#to check the percentage of churn based on the duration status
df.groupby(['DurationStatus',"Churn"]).size()/tot_data*100
```

## Summary
From analysis that we have done, there are some characteristics that have a significance of them being churn, such as:

1. User with `PaymentMehtod` of **Electronic check**.
2. User with `MonthlyCharges` **above** the average.
3. User with `DurationStatus` **very low**.

### Fitler and save data of potential churn customer to excel

```python
#to filter the dataframe of our potentioal churn customer
df_potential_churn = df[
    (df["PaymentMethod"]=='Electronic check')& \
    (df["MonthlyChargesStatus"]=='above')& \
    (df["DurationStatus"]=='very low')
].iloc[:, 0:6] #other method: ['customerID', 'gender', 'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn']
```
